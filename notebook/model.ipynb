```python
import os
import pandas as pd
import yt_dlp

# Load dataset CSV
df = pd.read_csv("dataset.csv")  # ganti sesuai nama file kamu

# Folder dataset
os.makedirs("dataset", exist_ok=True)

# yt-dlp options
ydl_opts = {
    'format': 'mp4',
    'outtmpl': 'dataset/%(id)s.%(ext)s',  # save dengan id video
    'quiet': False,
}


```

```python
# Download loop
with yt_dlp.YoutubeDL(ydl_opts) as ydl:
    for idx, row in df.iterrows():
        url = row['video']
        label = row['emotion']
        video_id = row['id']

        # buat folder label (Surprise, Trust, dll.)
        label_folder = os.path.join("dataset", label)
        os.makedirs(label_folder, exist_ok=True)

        try:
            # simpan di folder label dengan nama id.mp4
            ydl.download([url])
            downloaded_file = f"dataset/{video_id}.mp4"
            new_path = os.path.join(label_folder, f"{video_id}.mp4")
            if os.path.exists(downloaded_file):
                os.rename(downloaded_file, new_path)
            print(f"✅ Download sukses: {new_path}")
        except Exception as e:
            print(f"❌ Gagal download {url}: {e}")
```
